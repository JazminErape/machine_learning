```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
titanic_data = pd.read_csv('../data/titanic.csv', index_col='PassengerId')
titanic_data.head()
```

```python
titanic_data[titanic_data.Age.isnull()]
titanic_data[titanic_data.Cabin.isnull()]
titanic_data[titanic_data.Embarked.isnull()]
```

```python
fig = plt.figure()
pclass_age_notnull = titanic_data.Pclass[titanic_data.Age.notnull()]
(pclass_age_notnull).hist()
fig = plt.figure()
pclass_age_null = titanic_data.Pclass[titanic_data.Age.isnull()]
(pclass_age_null).hist()
```

```python
titanic_data.isnull()
have_null = pd.Series([False]*len(titanic_data), index=titanic_data.index)
for column in titanic_data.columns:
    have_null = have_null | titanic_data[column].isnull()
```

```python
titanic_data[have_null.apply(lambda x: not x)]
```

```python
titanic_data.dropna()
```

```python
age_mean = titanic_data.Age.mean()
```

```python
titanic_data.loc[titanic_data.Age.isnull(), 'Age'] = age_mean
```

```python
titanic_data['Cabin'] = titanic_data.Cabin.apply(lambda c: c[0] if type(c) == str else 'X')
```
